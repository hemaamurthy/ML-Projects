# Visual Analysis

```python
#Pie Chart: Visualizes the percentage of fraud vs. non-fraud transactions.
```

```python
class_percentages = (credit['Class']).value_counts() / len(credit) * 100

# to create a pie chart
plt.figure(figsize=(5, 5))  # describes the fiqure size
plt.pie(class_percentages, labels=class_percentages.index, autopct="%.1f%%",startangle=140)
plt.title('Class Percentage') #title
plt.axis('equal')

# to save the plot as an image
plt.savefig('class_percentage.png', bbox_inches='tight')
plt.show()
```

```python
# A bar graph can visually represent the distribution of the classes more clearly,
#especially when dealing with imbalanced datasets.
#It can help in visualizing the disparities between the classes more effectively than a pie chart.
```

```python
#Bar Chart: Shows the count of fraud and non-fraud transactions.
```

```python
sns.catplot(x="Class",kind='count',data=credit,height= 5)
plt.title("Class Count",fontsize=18)
plt.xlabel("fraud",fontsize=15)
plt.ylabel("count",fontsize=15)
plt.show()
```

```python
timedelta = pd.to_timedelta(credit['Time'],unit='s')
credit['Time_min'] = (timedelta.dt.components.minutes).astype(int)
credit['Time_hour'] = (timedelta.dt.components.hours).astype(int)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

# Time analysis

```python
#examining the time distribution by hour for both types of data 
#can reveal if there are specific hours or time periods when fraudulent activities are more likely to occur.
#This detailed analysis can enhance the understanding of the data
#and potentially improve the fraud detection capabilities of the model.
```

```python
#By Hours: Compares the distribution of transaction times between fraud and non-fraud transactions by the hour of the day.
```

```python
plt.figure(figsize=(12,5))
sns.distplot(credit[credit['Class'] == 0] ["Time_hour"],color='g')
sns.distplot(credit[credit['Class'] == 1] ["Time_hour"],color='r')
plt.title('Fraud x Normal Transaction by Hours', fontsize = 17)
plt.xlim([-1,25])
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.distplot(credit[credit['Class'] == 0] ["Time_min"],color='g')
sns.distplot(credit[credit['Class'] == 1] ["Time_min"],color='r')
plt.title('Fraud x Normal Transaction by Minutes',fontsize = 17)
plt.xlim([-1,61])
plt.show()
```

```python
#By Minutes: Compares the distribution of transaction times between fraud and non-fraud transactions by the minute within an hour.
```

# Transaction Amount Analysis

```python
# descriptive statistics for fraud and non fraud transactions
fraud = credit[credit['Class'] == 1]
normal = credit[credit['Class'] == 0]

print("fraud transaction statistics")
print(fraud["Amount"].describe())
print("\nnormal transaction statistics")
print(normal["Amount"].describe())
```

```python
plt.figure(figsize=(14,6))
plt.subplot(121)
ax = sns.boxplot(x="Class", y="Amount",data=credit)
ax.set_title("Class x Amount", fontsize=20)
ax.set_xlabel("Fraud",fontsize=16)
ax.set_ylabel("Amount(US)",fontsize=16)

plt.subplot(122)
ax = sns.boxplot(x="Class", y=np.log1p(credit["Amount"]),data=credit)
ax.set_title("Class x Amount", fontsize=20)
ax.set_xlabel("Fraud",fontsize=16)
ax.set_ylabel("Amount(Log)",fontsize=16)   
 
plt.subplots_adjust(hspace = 0.6,top = 0.8)
plt.show()                 
                 
```

### Left Plot (Amount in USD):

Non-fraudulent transactions (class 0) have a wider range of amounts, including many high-value outliers.
Fraudulent transactions (class 1) generally have lower amounts with fewer high-value outliers.

### Right Plot (Amount in Log Scale):

When viewed on a log scale, fraudulent transactions have a slightly lower median amount but similar overall distribution to non-fraudulent transactions.
This transformation reduces the impact of extreme values, highlighting that the distribution of transaction amounts is somewhat similar, though with differences in the range.

```python
ax = sns.lmplot(y="Amount",x="Time_min",fit_reg=False,aspect=1.8,data=credit,hue='Class')
plt.title("Amounts by Minutes of Frauds and NormalTransactions",fontsize=16)
plt.show()
```

#### Fraudulent transactions (orange dots) occur across all time periods but with lower amounts compared to non-fraudulent transactions (blue dots).
Non-fraudulent transactions exhibit a wide range of amounts, with some very high values, while fraudulent transactions are generally lower in amount.

# Correlation Matrix

#### Correlation in this context means how much two things in the 'credit' dataset are connected or related to each other. The heatmap you see helps show these connections visually by using colors to represent the strength and direction of the relationships between different parts of the data

```python
plt.figure(figsize=(12,8))
corr = credit.corr()
sns.heatmap(corr, cmap='PRGn',annot=False,fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

#### The correlation matrix shows that most variables have low correlations with each other, indicating they are largely independent.
